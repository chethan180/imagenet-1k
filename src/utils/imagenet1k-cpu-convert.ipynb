# Upload the checkpoint file here before running the script as best_resnet50_imagenet1k.pth

```python
import torch
from torchvision.models import resnet50

# Initialize model
model = resnet50(num_classes=1000)  # Since you trained on ImageNet-1K

# Load checkpoint
checkpoint = torch.load("best_resnet50_imagenet1k.pth", map_location="cpu")

# If the checkpoint has a 'state_dict' key, extract it
if "state_dict" in checkpoint:
    state_dict = checkpoint["state_dict"]
else:
    state_dict = checkpoint  # plain state_dict

# Some training scripts save keys like "module.conv1.weight" (for DataParallel)
# Let's clean that up if necessary
new_state_dict = {}
for k, v in state_dict.items():
    if k.startswith("module."):
        k = k[len("module."):]
    new_state_dict[k] = v

# Load the cleaned state dict
model.load_state_dict(new_state_dict, strict=True)

# Convert to CPU-only version (for Hugging Face)
model.cpu()

# Save it as CPU-only version
torch.save(model.state_dict(), "resnet50_imagenet1k_cpu.pth")

print("âœ… Model successfully converted to CPU version!")
```
